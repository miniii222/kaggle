# variables plots

## scatter plot

```python
def featrue_scatter(df1, df2, features) :
    i = 0
    sns.set_style('whitegrid')
    plt.figure()
    fig, ax = plt.subplots(4,4, figsize = [14,14])
    
    for fe in features :
        i+=1
        plt.subplot(4,4,i)
        plt.scatter(df1[fe], df2[fe], marker = '+')
        plt.xlabel(fe)
        
    plt.show()
    
```

```python
#train[::20] : start=0, stop=200000, step=20
featrue_scatter(train[::20], test[::20], features[:16])
```
